```python
# Import dependencies
import matplotlib.pyplot as plt
import geopandas as gpd
import earthpy as et
import contextily as ctx
%matplotlib inline
```

```python
# Load the shape data
save_path = "path/to/shapefile"
amz_shp = gpd.read_file(save_path)
```

```python
# Plot the shape data
amz = amz_shp.to_crs(epsg=3857)
ax = amz.plot(figsize=(10,10))
ctx.add_basemap(ax, url=ctx.providers.Stamen.Terrain)
ax.set_axis_off()
plt.show()
```
